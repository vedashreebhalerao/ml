```python
print(np.where(df['age']>78))#prints the index where the age greater than 78 
```

```python
sorted_df = df.sort_values(by=['age'], ascending=True)

Q1 = np.percentile(sorted_df['age'],25)
Q3 = np.percentile(sorted_df['age'],75)

IQR = Q3 - Q1
print(IQR)
```

```python
lower_bound = Q1 - (1.5*IQR)
upper_bound = Q3 + (1.5*IQR)

print("Min: ", lower_bound)
print("Max: ", upper_bound)
```

```python
sorted_df
```

```python
outliers = []

for i in df['age']:
    if(i<lower_bound or i > upper_bound):
        outliers.append(i)
    

print(len(outliers))
print(outliers)
```

Hangling Outliers

1.Removing
2.quartile based flooring and cappin
3.mean/median imputation


```python
#medina way to handling outliers
median = np.median(df['age'])
print(median)
for i in outliers:
    df['age'] = np.where(df['age']==i,37,df['age']) #37 = median
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42, shuffle=True)

```

```python
print('X training shape: ', X_train.shape)
print('X test shape: ', X_test.shape)
print('y training shape: ', y_train.shape)
print('y test shape: ', y_test.shape)
```

```python
#now we scale the dataset, we try to bring the mean similar to each other - DATA SCALING, a single value cant overpower the smaller values


#standard scaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)# scaling of mean of the 80% training set
X_test_std = scaler.transform(X_test) # scaling of mean by on old mean as we only have 20% of records for testing  (bigger chunks of data with big range)
```
